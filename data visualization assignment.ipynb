```python
from sklearn import decomposition
```

```python
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
X=iris.data[:,0:3]
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
label_iris_flower=LabelEncoder()
```

```python
y=iris.target
```

```python
y_label=label_iris_flower.fit_transform(y)
```

```python
label_iris_flower.classes_
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
pca=decomposition.PCA(n_components=3)
```

```python
X_trasform=pca.fit_transform(X_scaled)
```

```python
X_trasform[:5]
```

```python
fig=plt.figure(1,figsize=(10,8))
plt.clf()
ax=Axes3D(fig,rect=[0,0,0.95,1])
plt.cla()
ax.scatter(X_trasform[:,0],X_trasform[:,1],X_trasform[:,2],c=y,edgecolor='b')
plt.show()
```
